```python
def castfn(x):
  if isinstance(x, str):
    x = x.replace(' ','')
    x = x.replace(',','')
    x = x.replace('-','')
  try:
    return int(x)
  except:
    if len(x) ==0:
      print('x0',x)
      return -1
    print('xx',x)
    return -1
```

```python
house_price['분양가격'] = house_price['분양가격'].fillna(-1)
```

```python
house_price['분양가격'] = house_price['분양가격'].apply(castfn)
```
